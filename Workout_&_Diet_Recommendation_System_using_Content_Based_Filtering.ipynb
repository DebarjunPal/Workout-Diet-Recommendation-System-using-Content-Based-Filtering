```python
from sklearn.metrics.pairwise import cosine_similarity
import random
# pandas is already imported in an earlier cell. Assuming `data` and `scaler` are globally available.

def get_recommendation(top_n = 3):
    print("Please Enter your details for a personalized workout and recommendation: ")
    user_input = {
        'Sex': int(input("Enter Gender (Male: 1/ Female: 0): ")),
        'Age': int(input("Enter Age: ")),
        'Height': float(input("Enter Height: ")),
        'Weight': float(input("Enter Weight: ")),
        'Hypertension': int(input("Do you have hypertension? (Yes: 1/ No: 0): ")),
        'Diabetes': input("Do you have diabetes? (Yes/No): "), # Changed to string input
        'BMI': float(input("Enter BMI: ")),
        'Level': int(input("Enter Level (underweight: 3, Normal: 0, Overweight: 2, Obese: 1): ")),
        'Fitness Goal': int(input("Enter Fitness Goal (Weight Gain: 0, Weight Loss: 1): ")),
        'Fitness Type': int(input("Enter Fitness Type (Muscular Fitness: 1, Cardio Fitness: 0): ")) # Adjusted prompt
    }

    # Normalizing the Numerical Inputs:
    num_features = ['Age', 'Height', 'Weight', 'BMI']
    user_numeric_df = pd.DataFrame([user_input], columns = num_features)
    # The global `scaler` is used here
    user_numeric_scaled = scaler.transform(user_numeric_df[num_features])

    # Update user_input with scaled values
    for i, feature in enumerate(num_features):
        user_input[feature] = user_numeric_scaled[0, i]

    # Create a DataFrame for the user's input, dropping 'Diabetes' for similarity calculation
    # as it's not encoded in the main `data` DataFrame and needs to be excluded.
    features_for_similarity = ['Sex', 'Age', 'Height', 'Weight', 'Hypertension', 'BMI', 'Level', 'Fitness Goal', 'Fitness Type']
    user_features_df = pd.DataFrame([user_input])[features_for_similarity]

    # Prepare `data` DataFrame for similarity calculation (only relevant features)
    data_features_df = data[features_for_similarity]

    # Calculating Similarity Scores
    # Calculate similarity between user's input features and all data features
    similarity_scores = cosine_similarity(user_features_df, data_features_df)
    similarity_scores_flat = similarity_scores.flatten() # Flatten for easier sorting

    # Retrieving Top similar users and getting the first recommendation
    # Get indices of top_n most similar users
    similar_user_indices = similarity_scores_flat.argsort()[-top_n:][::-1]
    similar_users = data.iloc[similar_user_indices]
    recommendation_1 = similar_users[['Exercises', 'Diet', 'Equipment']].iloc[0] # Corrected .node() to .iloc[0]

    simulated_recommendations = []
    feedback_matrix = [] # Initialize feedback_matrix

    # Generate additional recommendations (top_n - 1)
    for _ in range(top_n - 1):
        modified_input = user_input.copy()
        # Randomly adjusting Age, Weight and BMI (these are already scaled values)
        # A more robust approach would be to inverse_transform, modify raw, then transform.
        modified_input['Age'] += random.uniform(-0.5, 0.5) # Use float for scaled values
        modified_input['Weight'] += random.uniform(-0.5, 0.5)
        modified_input['BMI'] += random.uniform(-0.2, 0.2)

        # Create DataFrame for modified input and filter features
        modified_user_features_df = pd.DataFrame([modified_input])[features_for_similarity]

        # Calculating similarity scores for modified input
        modified_similarity_scores = cosine_similarity(modified_user_features_df, data_features_df).flatten()
        # Get the single most similar user for this modified input
        modified_similar_user_index = modified_similarity_scores.argsort()[-1:][::-1][0]
        recommendation = data.iloc[modified_similar_user_index][['Exercises', 'Diet', 'Equipment']]

        # Check if the recommendation is already in simulated_recommendations or recommendation_1
        is_duplicate = False
        for rec_existing in simulated_recommendations + [recommendation_1]:
            if rec_existing.equals(recommendation):
                is_duplicate = True
                break
        if not is_duplicate:
            simulated_recommendations.append(recommendation)
        # If it's a duplicate, we just don't add it, so we might end up with fewer than top_n - 1 simulated recs if duplicates are common.

    print("\nRecommended Workout and Diet Plans based on your input: ")
    print("\nRecommendation 1: ")
    print("Exercises: ", recommendation_1["Exercises"]) # Corrected spelling
    print("Diet: ", recommendation_1["Diet"]) # Added colon
    print("Equipment: ", recommendation_1["Equipment"]) # Corrected spelling

    for idx, rec in enumerate(simulated_recommendations, start=2):
        print(f"\nRecommendation {idx}: (Slight Variation)")
        print("Exercises: ", rec["Exercises"]) # Corrected spelling
        print("Diet: ", rec["Diet"]) # Added colon
        print("Equipment: ", rec["Equipment"]) # Corrected spelling

    # Collecting feedback for each recommendation
    all_recommendations = [recommendation_1] + simulated_recommendations
    for i in range(len(all_recommendations)):
        feedback = int(input(f"Was recommendation {i+1} relevant? (Yes: 1, No: 0): "))
        feedback_matrix.append(feedback)

    # Calculating MRR(Mean Reciprocal Rank)
    relevant_indices = [i+1 for i, feedback in enumerate(feedback_matrix) if feedback == 1]
    if relevant_indices:
        mrr = np.mean([1/rank for rank in relevant_indices])
    else:
        mrr = 0.0
    print(f"\nMean Reciprocal Rank (MRR): {mrr}")

    return all_recommendations

# Getting the recommendations
recommendations = get_recommendation(top_n = 3)
```
